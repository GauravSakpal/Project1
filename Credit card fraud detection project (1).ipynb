```python
#Splitting the data into Features & Targets
X=new_data.drop(columns="Class",axis=1)
Y=new_data["Class"]
```

```python
X
```

```python
Y
```

```python
#Split the data into Training data & Testing Data

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
#Model Training

#Logistic Regression

model=LogisticRegression()
model.fit(X_train, Y_train)
```

```python
#Model Evaluation

#Accuracy Score

# accuracy on training data
X_train_pred=model.predict(X_train)
train_data_accuracy=accuracy_score(Y_train, X_train_pred )
print(train_data_accuracy)
```

```python
# accuracy on test data
X_test_pred=model.predict(X_test)
test_data_accuracy=accuracy_score(Y_test, X_test_pred)
print(test_data_accuracy)
```

```python
#Confusion metrix
from sklearn.metrics import confusion_matrix
```

```python
cf_matrix_test=confusion_matrix( Y_test, X_test_pred)
print(cf_matrix_test)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.heatmap(cf_matrix_test, annot=True)
plt.xlabel("Predicated values")
plt.ylabel("True values")
plt.title("Confussion matrix")
```

```python
#Precision
from sklearn.metrics import precision_score

precision_score_test=precision_score(Y_test, X_test_pred)
print(precision_score_test)
```

```python
#Recall
from sklearn.metrics import recall_score

recall_score_test=recall_score(Y_test, X_test_pred)
print(recall_score_test)
```

```python
#F1 Score
from sklearn.metrics import f1_score

f1_score_test=f1_score(Y_test, X_test_pred)
print(f1_score_test)
```

```python
def precision_recall_f1_score(true_labels, pred_labels):
    precision_value = precision_score(true_labels, pred_labels)
    recall_value = recall_score(true_labels, pred_labels)
    f1_score_value = f1_score(true_labels, pred_labels)

    print('Precision =',precision_value)
    print('Recall =',recall_value)
    print('F1 Score =',f1_score_value)
     

  
```

```python
# classification metrics for test data
precision_recall_f1_score(Y_test , X_test_pred)
```

```python
# classification metrics for train data
precision_recall_f1_score(Y_train , X_train_pred)
```
